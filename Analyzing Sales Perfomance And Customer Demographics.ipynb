```python
df1['Date']=pd.to_datetime(df1['Date'],dayfirst=True)
df1
```

```python
DAY_SALES=df1.groupby('Date').apply(lambda x: (x['Quantity']*x['Price']).sum()).reset_index(name='Total_Sales')
DAY_SALES
```

```python
PEAK_DAY_SALES=DAY_SALES.sort_values(by='Total_Sales',ascending=False).head(5)
```

```python
df1['Promotion']=np.random.choice([0,1],size=len(df1))
df1['Promotion']
df1
```

```python
daily_sales_promo=df1.groupby(['Date','Promotion']).apply(lambda x: (x['Quantity']*x['Price']).sum()).reset_index(name='Total Sales')
```
